```python
def get_prec_recall(cl_reports):
    precs = []
    recs = []
    total_neg_cl = []
    for cl_report in cl_reports:
        prec = float(cl_report.split('\n')[3].split()[1])
        rec = float(cl_report.split('\n')[3].split()[2])
        precs.append(prec)
        recs.append(rec)
        total_neg_cl.append(int(cl_report.split('\n')[3].split()[4]))
    return (precs, recs, total_neg_cl)
```

```python
def plot_class_report_negative_class(df, title):
    pp_steps_name_map = {
            'raw' : 'Raw',  
            'convert_to_lower' : 'To_lower',
            'remove_whitespaces' : 'Remove_whitespaces', 
            'remove_leaky' : 'Remove_leaky' , 
            'replace_abbreviation_words' : 'Replace_abbreviations',
            'strip_non_printable_chars' : 'Strip_non_printables',
            'replace_acronyms' : 'Replace_acronyms',
            'remove_stopwords': 'Remove_stopwords',
            'remove_rare_words' : 'Remove_rare_words',
            'remove_non_alphanumeric' : 'Remove_non_alphanumeric_chars',
            'remove_non_alphabet_words' : 'Remove_non_alphabet_chars',
            'remove_words_containing_non_alphabets' : 'Remove_words_containing_non_alpha',
            'black_listed_words_regex_mapping' : 'Regex_maping_black_list',
            'check_if_proper_name_place_or_ethnicity' : 'Check_if_name',
            'replace_profane_words_using_fuzzy' : 'Fuzzy_profane_mapping',
            'replace_common_words_using_fuzzy' : 'Fuzzy_common_words_mapping',
            'lemmatize_english_words' : 'Lemmatize',
            'stemming_english_words' : 'Stemming',
            'extract_info_from_url' : 'URL_info_extract',
            'remove_whitespaces remove_leaky replace_abbreviation_words replace_acronyms' : 'Preprocess_order-1', 
            'remove_whitespaces remove_leaky replace_abbreviation_words replace_acronyms remove_stopwords remove_rare_words' : 'Preprocess_order-2',
            'remove_whitespaces remove_leaky replace_abbreviation_words replace_acronyms black_listed_words_regex_mapping replace_profane_words_using_fuzzy replace_common_words_using_fuzzy'  : 'Preprocess_order-3',
            'remove_whitespaces remove_leaky convert_to_lower remove_stopwords remove_rare_words' : 'Preprocess_order-4'   
            }
    df.index= pp_steps_name_map.values()
    ymax = df.max().max()
    if ymax <=1 :
        ymax = 1
        ymin = df.min().min() - 0.1
    else: 
        ymax = ymax + 100
        ymin = df.min().min() - 100

    ax = df.plot(ylim=(ymin,ymax), figsize=(15, 10), rot=45, marker='*', title=title)
    ax.set_xticks(range(len(df.index)))
    ax.set_xticklabels([item for item in df.index.tolist()], rotation=75);
```

```python
prec_df = []
recall_df = []
total_neg_cl_df = []
for algo in [logit, nbsvm, fasttext, xgboost]:
    precs, recs, total_neg_cl = get_prec_recall(algo.class_report.values)
    prec_df.append(precs)
    recall_df.append(recs)
    total_neg_cl_df.append(total_neg_cl)
```
